# WEATHER ANALYSIS

ANALYSIS

* The weather is warmer when closer to the equator as seen in the Latitude vs Max Temperature scatterplot, as the latitude increases the temperature starts dropping.
* The humidity and latitude do not show any patterns or strong relationships as the scatterpoints are pretty much scattered all over with no observable impact of each other on the cities weather but maximum number of cities tend to have a humidity between 60 - 100% .
* The city latitude and cloudiness also have an unrelated pattern and most cities receive cloudiness everywhere on the latitude. The windspeed for most cities remains between 1 to 10 mph all over the latitude with no observable patterns or trends indicating a near even distribution.


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from config_2 import weather_api_key


# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy


# Output File (CSV)
output_data_file = "output_data/cities.csv"


# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

```

```python
#Generate Cities List
# List for holding lat_lngs and cities

lat_lngs = []
cities = []




# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)






# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    
    
    
    
    

# If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)


        

        

# Print the city count to confirm sufficient count
print(len(cities))
print(city)
```

```python
#Perform API Calls
#Perform a weather check on each city using a series of successive API calls.
#Include a print log of each city as it's being processed (with the city number and city name).

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID="+weather_api_key
print(url)
```

```python
cit = []
country= []
lat_list = []
long_list = []
cloudiness = []
humidity = []
max_temp= []
windspeed = []
date = []
counter = 0 


print("starting data retreival")

for city in cities:
    url2 = url + "&q="+city
    #q = query
    counter += 1  #so we could start with call no 1 in the print statement
    print(f" Api call no  {counter} set 1 ; {city}")
    print(url2)
    
    response = requests.get( url2 ).json()
    if response["cod"]==200: 
        
        cit.append(response["name"])
        country.append(response["sys"]["country"])
        lat_list.append(response["coord"]["lat"])
        long_list.append(response["coord"]["lon"])
        cloudiness.append(response["clouds"]["all"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        windspeed.append(response["wind"]["speed"])
        date.append(response["dt"])
    
    else:
        cit.append("na")
        country.append("na")
        lat_list.append("na")
        long_list.append("na")
        cloudiness.append("na")
        humidity.append("na")
        max_temp.append("na")
        windspeed.append("na")
        date.append("na")
    time.sleep(1)# means pause for a second
    
    
print("DATA RETREIVAL COMPLETE")
    
```

```python
import pandas as pd

#Convert Raw Data to DataFrame
weather_df = pd.DataFrame({"City": cit,
                           "Cloudiness": cloudiness,
                           "Country":country,
                           "Date": date,
                           "Humidity": humidity,
                           "Latitude": lat_list,
                           "Longitude": long_list,
                           "Max Temperature": max_temp,
                           "Windspeed": windspeed,
                           
    
    
})


#Export the data into a csv file
weather_df.to_csv('Weather_csv', index=False , header= True)

len(weather_df)
```

```python
weather_df2 = weather_df[weather_df.City !="na"]
weather_df2.head(10)
```

# Latitude vs. Temperature Plot  

```python
import matplotlib.pyplot as plt
import datetime

now = datetime.datetime.now()

plt.scatter(weather_df2["Latitude"],weather_df2["Max Temperature"],marker="o",facecolors="purple",edgecolors="black")

plt.title("City Latitude vs Max Temp  " + ("(" + str(now.month)) + "-"+ (str(now.day)) +"-"+ (str(now.year))+ ")")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.savefig("latvstemp.png", bbox_inches = "tight")
plt.show()
```
